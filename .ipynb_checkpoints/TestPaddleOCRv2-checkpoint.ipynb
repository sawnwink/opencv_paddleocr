```python
from paddleocr import PaddleOCR,draw_ocr
import numpy as np
from matplotlib import pyplot as plt
import cv2
from PIL import Image

# Define our imshow function 
def imshow(title = "Image", image = None, size = 10):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
    
    
# Paddleocr supports Chinese, English, French, German, Korean and Japanese.
# You can set the parameter `lang` as `ch`, `en`, `french`, `german`, `korean`, `japan`
# to switch the language model in order.
# ocr = PaddleOCR(use_angle_cls=True, lang='en') # need to run only once to download and load model into memory
# ocr = PaddleOCR(use_angle_cls=True)
# img_path = 'images/ar1.jpg'

img_path = 'images/image.jpg'

ocr = PaddleOCR(use_angle_cls=True, lang='ch') 

result = ocr.ocr(img_path, cls=True)
# for line in result:
#     print(line)

# ---------- draw result (read with PIL, display with PIL)---------------
# from PIL import Image
# image = Image.open(img_path).convert('RGB')

# boxes = [line[0] for line in result]
# txts = [line[1][0] for line in result]
# scores = [line[1][1] for line in result]

# im_show = draw_ocr(image, boxes, txts, scores, font_path='doc/fonts/simfang.ttf')
# im_show = Image.fromarray(im_show)
# im_show.save('result.jpg')

# img = cv2.imread('result.jpg')
# imshow("Output Image", img, 100)


#-----draw result (Read with OpenCV, Display with PIL)----------
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# need to convert to RGB because
    
boxes = [line[0] for line in result]
txts = [line[1][0] for line in result]
scores = [line[1][1] for line in result]

# for box in boxes:
# #     [[40.0, 22.0], [234.0, 21.0], [234.0, 51.0], [40.0, 52.0]]
#     print(box)
#     cv2.rectangle(img,(int(box[0][0]),int(box[0][1])),(int(box[2][0]),int(box[2][1])),(0,255,0),1)
im_show = draw_ocr(img, boxes, txts, scores, font_path='doc/fonts/simfang.ttf')
# im_show = Image.fromarray(im_show)
# im_show.save('result.jpg')
# from matplotlib import cm
# imgPIL = Image.fromarray(np.uint8(cm.gist_earth(img)*255))

# 
# imgPIL = Image.fromarray(im_show)
# imgPIL=imgPIL.convert('RGB')

im_show = cv2.cvtColor(im_show, cv2.COLOR_BGR2RGB)
imshow("Output Image", im_show, 100)

```
